```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# 1. Crear el diccionario con tus datos de la tabla de LaTeX
data = {
    "Modelo": [
        "Regresión Ridge",
        "SARIMA",
        "Prophet",
        "XGBoost",
        "LightGBM",
        "LSTM",
        "GRU",
        "LSTM_logdiff",
    ],
    "Familia 106": [
        "106,26",
        "107,29",
        "117,10",
        "113,38",
        "118,23",
        "108,97",
        "108,93",
        "117,85",
    ],
    "Familia 124": [
        "70,05",
        "64,93",
        "41,87",
        "65,27",
        "56,06",
        "65,63",
        "73,44",
        "31,96",
    ],
    "Familia 233": [
        "70,69",
        "59,86",
        "85,15",
        "43,18",
        "45,74",
        "45,12",
        "54,24",
        "53,31",
    ],
}

# 2. Convertir a DataFrame y establecer los Modelos como el índice
df = pd.DataFrame(data)
df.set_index("Modelo", inplace=True)

# 3. Limpiar los datos: cambiar comas por puntos y convertir a tipo numérico (float)
for col in df.columns:
    df[col] = df[col].str.replace(",", ".").astype(float)

# 4. Configurar el estilo del gráfico
plt.figure(figsize=(8, 6))

# Usamos la paleta Red-Yellow-Green invertida
sns.heatmap(
    df,
    annot=True,  # Muestra los valores numéricos dentro de cada celda
    fmt=".2f",  # Formato con 2 decimales
    cmap="RdYlGn_r",  # Escala de colores (Verde = Mejor, Rojo = Peor)
    linewidths=0.5,  # Línea de separación entre celdas
    cbar_kws={"label": "SMAPE (%)"},  # Etiqueta de la barra de color
)

# 5. Añadir títulos y etiquetas
plt.title(
    "SMAPE (%) comparativo por modelo y familia de producto",
    fontsize=14,
    pad=15,
)
plt.ylabel("Modelo", fontsize=12)
plt.xlabel("Familia de Producto", fontsize=12)

# Guardar la imagen en formato vectorial PDF en la ruta especificada
plt.savefig("../figuras/07_heatmap_global.pdf", bbox_inches="tight")

plt.show()
```
